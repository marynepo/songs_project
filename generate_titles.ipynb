```python
import torch
import sqlite3
import pandas as pd
from simpletransformers.t5 import T5Model
from sklearn.model_selection import train_test_split

conn = sqlite3.connect('songs.db')
c = conn.cursor()
```

```python
df = pd.read_sql('SELECT * FROM Splitted', conn)
df1 = df[['genre','full_lyrics', 'full_chorus', 'full_verses']].copy()
X_train, X_test, y_train, y_test = train_test_split(df1, df['title'], stratify = df1['genre'], test_size=0.15, random_state=42) 
```

```python
use_cuda = torch.cuda.is_available()
```

```python
def count_matches(labels, preds):
    print(labels)
    print(preds)
    return sum([1 if label == pred else 0 for label, pred in zip(labels, preds)])


def title_results(col):
  train_df = pd.DataFrame(list(zip(['generate title'] * len(y_train), X_train[col].values, y_train)), columns=["prefix", "input_text", "target_text"])
  eval_df = pd.DataFrame(list(zip(['generate title'] * len(y_test), X_test[col].values, y_test)), columns=["prefix", "input_text", "target_text"])
  model_args = {
    "max_length": 17,
    "reprocess_input_data": True,
    "overwrite_output_dir":True,
    "train_batch_size": 8,
    "length_penalty": 20,
    "repetition-penalty": 15,
    "num_train_epochs": 5,
    "save_eval_checkpoints": False,
    "save_model_every_epoch": False,
    "silent": False,
    "evaluate_generated_text": True,
    "evaluate_during_training": False,
    "evaluate_during_training_verbose": False,
    "use_multiprocessing": False,
    "save_best_model": True,
    "output_dir": './en_output/' + col + '/',
    "preprocess_inputs": True
  }
  model = T5Model("t5", "t5-base", args=model_args, use_cuda=use_cuda)
  model.train_model(train_df, eval_data=eval_df, matches=count_matches)
  with torch.no_grad():
    print(model.eval_model(eval_df, matches=count_matches))
```

```python
title_results('full_lyrics')
```

```python
title_results('full_chorus')
```

```python
title_results('full_verses')
```
